```python
csvfile = open('Call Center All.csv', 'w', encoding='utf-8')
writer = csv.writer(csvfile)
writer.writerow(['job title', 'company name', 'location', 'timetype_Skills'])

for i in range(30):
    if i == 0:
        Url = "https://wuzzuf.net/search/jobs/?a=hpb%7Cspbg&q=call%20center"
    else:
        url = f"https://wuzzuf.net/search/jobs/?q=call+center&a=hpb&start={i}"
        website = requests.get(url).text
        website = BeautifulSoup(website , "lxml")
        jobs = website.find_all("div", class_ = "css-pkv5jc")
        for job in jobs:
            jobtitle = job.find("a", class_ = "css-o171kl").text
            companyname = job.find("a", class_ = "css-17s97q8").text
            location = job.find("span", class_ = "css-5wys0k").text
            timetype_Skills = job.find("div", class_ = "css-y4udm8").text
            # newdate = job.find("div", class_ = "css-4c4ojb").text
            writer.writerow([jobtitle, companyname, location, timetype_Skills])
csvfile.close()

```
