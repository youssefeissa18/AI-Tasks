```python
with open("AI Engineer1.csv", "w", encoding='utf-8') as jobsAI:
    wr = csv.writer(jobsAI)
    wr.writerow(["Job Title", "Company Name","Location" ,"job Type", "Skills"])
    jobs = soup.find_all("div", class_ = "css-pkv5jc")
    for job in jobs: 
        jobtitle = job.find("a", class_ = "css-o171kl").text
        companyname= job.find("a", class_ ="css-17s97q8").text
        location = job.find("span", class_ = "css-5wys0k").text
        jobtype = job.find("span", class_= "css-1ve4b75 eoyjyou0").text
        Experince = job.find_all("a", class_ = ["css-5x9pm1"])
        e = []
        for i in range(len(Experince)):
            e.append(Experince[i].text)
        wr.writerow([jobtitle, companyname, location, jobtype, e])
```
